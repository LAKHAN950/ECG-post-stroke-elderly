```python
import pandas as pd
```

```python
df = pd.read_csv("subjects.csv")

# dropping columns
df = df.iloc[:,[0,1,3,4,12]]
df.head()
```

```python
# replaing values in rows in a column
df1 = df.copy()

## column: group
df['group'] = df['group'].map({"CONTROL" : 0, "STROKE":1})
## column: group2
df['group2'] = df['group2'].map({"CONTROL-HTN":0, "CONTROL-NTN":0, 
                                 "STROKE-HTN":1, "STROKE":1})
## column: DM Non-DM STROKE
df['DM Non-DM STROKE'] = df['DM Non-DM STROKE'].map({"Non-DM":0,"STROKE":1})

df.head()
```

```python
df = df.loc[df["completed_visit_status"] == "COMPLETED"]
df.head()
```

```python
df.to_csv("ml_labels.csv", index = False)
```

```python
for i in ["group", "group2", "DM Non-DM STROKE"]:
    print("Count for {}:".format(i))
    print("Non-Stroke: ", df[i].value_counts()[0])
    print("Stroke: ", df[i].value_counts()[1])
```
